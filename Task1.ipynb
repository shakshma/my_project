```python
from sklearn.linear_model import LinearRegression
x_train, y_train = train_data.drop(['dis'],axis=1), train_data['dis']
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
test_data=x_test.join(y_test)

test_data['rm']=np.log(test_data['rm'] + 1)
test_data['dis']=np.log(test_data['dis'] + 1)
test_data['rad']=np.log(test_data['rad'] + 1)
test_data['tax']=np.log(test_data['tax'] + 1)

test_data['medv']=test_data['dis']/test_data['tax']
test_data['ptratio']=test_data['rm']/test_data['age']
```

```python
x_test, y_test = test_data(['stat'],axis=1), test_data['stat']
```

```python
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor()
forest.fit(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Initialize the RandomForestRegressor
forest = RandomForestRegressor()

# Define the parameter grid
param_grid = {
    "n_estimators": [1, 10, 30],
    "max_features": [2, 4, 6, 8]
}

# Initialize GridSearchCV
gridsearch = GridSearchCV(forest, param_grid, cv=5,
                          scoring="neg_mean_squared_error",
                          return_train_score=True)

# Fit the model
gridsearch.fit(x_train, y_train)

# Display the best parameters and best score
print("Best parameters found: ", gridsearch.best_params_)
print("Best score: ", gridsearch.best_score_)
```
